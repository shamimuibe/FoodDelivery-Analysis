```python
# Function to check if any keyword is present in the text
def has_keyword(text, keywords):
    for keyword in keywords:
        if keyword in text:
            return 1
    return 0
```

```python
# Create new columns for Ease of Use, Intention to Use, and Usefulness
data['Ease_of_Use'] = data['content'].apply(lambda x: has_keyword(x, ease_of_use_keywords))
data['Intention_to_Use'] = data['content'].apply(lambda x: has_keyword(x, intention_to_use_keywords))
data['Usefulness'] = data['content'].apply(lambda x: has_keyword(x, usefulness_keywords))
```

```python
# Replace 1 in 'Ease_of_Use_Indicator' with -1 when 'sentiment' is -1
data['Ease_of_Use'] = data.apply(lambda row: -1 if row['sentiment'] == -1 else row['Ease_of_Use'], axis=1)
```

```python
data['Intention_to_Use'] = data.apply(lambda row: -1 if row['sentiment'] == -1 else row['Intention_to_Use'], axis=1)
```

```python
data['Usefulness'] = data.apply(lambda row: -1 if row['sentiment'] == -1 else row['Usefulness'], axis=1)
```

```python
import statsmodels.api as sm
```

```python
# Define the dependent and independent variables
X = data[['Ease_of_Use', 'Usefulness']]
y = data['Intention_to_Use']

#y = data['Intention_to_Use_Indicator']
```

```python
# Add a constant (intercept) to the independent variables
X = sm.add_constant(X)
```

```python
# Fit a linear regression model
model = sm.OLS(y, X).fit()
```

```python
# Print the regression summary
print(model.summary())
```
