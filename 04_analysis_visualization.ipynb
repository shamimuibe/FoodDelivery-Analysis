```python
import matplotlib.pyplot as plt
```

```python
# Extract the columns of interest
Ease_of_Use = df['Ease_of_Use']
Usefulness = df['Usefulness']
Intention_to_Use = df['Intention_to_Use']
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(Ease_of_Use, Intention_to_Use, label='Intention_to_Use vs. Ease_of_Use', marker='o', color='blue')
plt.scatter(Usefulness, Intention_to_Use, label='Intention_to_Use vs. Usefulness', marker='x', color='red')
```

```python
# Add labels and legend
plt.xlabel('Ease_of_Use / Usefulness')
plt.ylabel('Intention_to_Use')
plt.legend()
```

```python
# Show the plot
plt.title('Impact of Ease_of_Use and Usefulness on Intention_to_Use')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
#df = pd.read_csv('datafile.csv')

# Create a bar graph for each variable
variables = ['Ease_of_Use', 'Usefulness', 'Intention_to_Use']

for variable in variables:
    counts = df[variable].value_counts().sort_index()

    # Create a bar graph for the current variable
    plt.figure(figsize=(8, 6))
    counts.plot(kind='bar', color=['red', 'blue', 'green'])
    plt.title(f'Counts of {variable}')
    plt.xlabel(f'{variable} Values')
    plt.ylabel('Counts')
    plt.xticks(rotation=0)  # Set x-axis labels vertically for better readability
    plt.show()
```

```python
from wordcloud import WordCloud
```

```python
# Combine all text from the 'content' column into a single string
text = ' '.join(df['content'].astype(str))

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud for Content")
plt.show()
```
