```python
import pandas as pd
```

```python
# Load the three datasets
doordash_data = pd.read_csv('data/DoorDash.csv')
grubhub_data = pd.read_csv('data/Grubhub.csv')
uber_eats_data = pd.read_csv('data/Uber_Eats.csv')
```

```python
# Concatenate the datasets into one
fooddelivery_data = pd.concat([doordash_data, grubhub_data, uber_eats_data], ignore_index=True)
```

```python
# Save the merged dataset to a new CSV file
fooddelivery_data.to_csv('fooddelivery.csv', index=False)
```

```python
# Select and keep only the desired columns
desired_columns = ["content", "score", "thumbsUpCount", "Apps"]
fooddelivery_data = fooddelivery_data[desired_columns]
```

```python
# Save the updated dataset to a new CSV file
fooddelivery_data.to_csv('fooddelivery_updated.csv', index=False)
```
